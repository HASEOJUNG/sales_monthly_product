```python
import pandas as pd

df=pd.read_csv('./csv/train.csv')

df_transposed = df.set_index(['ID', '제품', '대분류', '중분류', '소분류', '브랜드']).transpose()

df_transposed.index = pd.to_datetime(df_transposed.index)

monthly_sum = df_transposed.groupby(df_transposed.index.to_period('M')).sum()

monthly_sum_transposed = monthly_sum.transpose().reset_index()

monthly_sum_transposed
```

```python
product_info=pd.read_csv('./csv/제공데이터/product_info.csv')

monthly_sum_transposed_info=pd.merge(monthly_sum_transposed, product_info, on='제품',how='inner')
monthly_sum_transposed_info
```

```python
df = monthly_sum_transposed_info.groupby(['대분류', '중분류', '소분류']).sum(numeric_only=True).reset_index()
df
```

```python
df['New_ID'] = df.index + 1
cols = ['New_ID'] + df.columns[:-1].tolist()
df = df[cols]
df
```

```python
df=df.drop('ID', axis=1)
df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

df.columns = df.columns.map(str)

monthly_columns = df.columns[4:]

plt.figure(figsize=(18, 10))

for idx, row in df.iterrows():
    plt.plot(monthly_columns, row[4:])

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales by Category')
plt.grid(True)
plt.tight_layout()
plt.gca().get_yaxis().set_major_formatter(FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```

```python
for col in df_not4.columns[3:]:
    df_not4[col] = pd.to_numeric(df_not4[col], errors='coerce')

    
columns_to_exclude = ['2023-03', '2023-04']
monthly_columns = [col for col in df_not4.columns[3:] if col not in columns_to_exclude]


monthly_columns = df_not4.columns[3:]

percentage_change = df_not4[monthly_columns].pct_change(axis=1)
threshold = 2
significant_changes = percentage_change[(percentage_change.abs() > threshold).any(axis=1)]

# Extract the IDs with significant changes
significant_changes_ids = significant_changes.index.tolist()

print("IDs with significant month-over-month changes:", significant_changes_ids)
```
