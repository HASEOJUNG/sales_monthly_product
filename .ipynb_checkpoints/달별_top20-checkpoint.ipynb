```python
import pandas as pd

df=pd.read_csv('./csv/제공데이터/train.csv')

df
```

```python
df_transposed = df.set_index(['ID', '제품', '대분류', '중분류', '소분류', '브랜드']).transpose()

df_transposed
```

```python
df_transposed.index = pd.to_datetime(df_transposed.index)
```

```python
monthly_sum = df_transposed.groupby(df_transposed.index.to_period('M')).sum()

monthly_sum
```

```python
monthly_sum_transposed = monthly_sum.transpose().reset_index()
monthly_sum_transposed
```

```python
product_info=pd.read_csv('./csv/제공데이터/product_info.csv')
```

```python
monthly_sum_transposed_info=pd.merge(monthly_sum_transposed, product_info, on='제품',how='inner')
monthly_sum_transposed_info

#5000개 정도가 제푸특성 없는 제품 
```

```python
monthly_sum_transposed_info2=pd.merge(monthly_sum_transposed, product_info, on='제품',how='outer')
monthly_sum_transposed_info2
```
