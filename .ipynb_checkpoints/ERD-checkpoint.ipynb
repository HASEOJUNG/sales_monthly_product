```python
df_melted = df.melt(id_vars=["ID", "제품", "대분류", "중분류", "소분류", "브랜드"], var_name="date", value_name="sales")

df_date_row = df_melted.groupby(["대분류", "중분류", "소분류", "브랜드","date"]).sum().reset_index()

df_date_row
```

```python
df_date_row_nobrand = df_melted.groupby(["대분류", "중분류", "소분류","date"]).sum().reset_index()
df_date_row_nobrand
```

```python
df_date_row_nobrand['date'] = pd.to_datetime(df_date_row_nobrand['date'])
df_date_row_nobrand
```

```python
df_date_row_nobrand['week']=df_date_row_nobrand['date'].dt.dayofweek
df_date_row_nobrand
```

```python
df_date_row_nobrand['ID'].unique()
```

```python
#ID 별로 sales 평균
average_sales_per_id = df_date_row_nobrand.groupby('ID')['sales'].mean()

#sales 높은 순으로 정렬
sorted_ids = average_sales_per_id.sort_values(ascending=False).index

for id in sorted_ids:
    print(f"ID: {id}, Average Sales: {average_sales_per_id[id]}")
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_date_row_nobrand, x='week', y='sales', hue='ID', palette='tab20', hue_order=sorted_ids, legend=None)

# for id, group_data in df_date_row_nobrand.groupby('ID'):
#     plt.text(6.2, group_data['sales'].iloc[-1], str(id), fontsize=8, color='black')


plt.xlabel('Week')
plt.ylabel('Sales')
plt.grid(True)
plt.xlim(0, 6) 
plt.show()
```

```python
df_date_row_nobrand['month']=df_date_row_nobrand['date'].dt.month
df_date_row_nobrand
```

```python
df_date_row_nobrand['month'].unique()
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=df_date_row_nobrand, x='month', y='sales', hue='ID', palette='tab20', hue_order=sorted_ids, legend=None)

# for id, group_data in df_date_row_nobrand.groupby('ID'):
#     plt.text(6.2, group_data['sales'].iloc[-1], str(id), fontsize=8, color='black')


plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.xlim(1,12) 
plt.show()
```
